```python
files = [
    "data/a-3kittens.txt",
    "data/bp-tom-kitten.txt",
    "data/el-owl-cat.txt",
    "data/rc-cat-fiddle.txt",
    "data/rk-cat.txt"
]

import pandas as pd
frequency_term_document_matrix =pd.DataFrame()

for file in files:
    with open(file) as f:
        raw_verse = f.read()
    f.close()

    verse = textblob.TextBlob(raw_verse)
    not_stop_words = [k for k in verse.words if k.lower() not in stopwords]
    verse = textblob.TextBlob(" ".join(not_stop_words))
    print(f"Words: {verse.words}")
    print(f"Words count: {verse.word_counts}")

    unique_words = sum(1 for value in verse.word_counts.values() if value == 1)
    print(f"Number of unique words: {unique_words}")
    highest_values = sorted(verse.word_counts.items(), key=lambda x: x[1], reverse=True)[:10]
    print("10 most frequent words:")
    for key, value in highest_values:
        if key not in frequency_term_document_matrix.columns:
            frequency_term_document_matrix[key] = 0
        print(f"{key}: {value}")
    highest_values = {key: value for key, value in highest_values}
    frequency_term_document_matrix = frequency_term_document_matrix.append(highest_values, ignore_index=True)
frequency_term_document_matrix.fillna(0, inplace=True)
boolean_term_document_matrix = frequency_term_document_matrix.copy()  # Copy the original DataFrame
boolean_term_document_matrix[boolean_term_document_matrix >= 1] = 1
boolean_term_document_matrix
```

```python
import numpy as np
from scripts.DistanceHelper import get_distance, DistanceType

number_of_rows = boolean_term_document_matrix.shape[0]
number_of_columns = boolean_term_document_matrix.shape[1]
euclidean_distance_matrix = np.zeros((number_of_rows, number_of_rows))

for i, element_0 in boolean_term_document_matrix.iterrows():
    for j, element_1 in boolean_term_document_matrix.iterrows():
        euclidean_distance_matrix[i, j] = get_distance(
            element_0, element_1, number_of_columns, DistanceType.EUCLIDEAN
        )
        
euclidean_distance_matrix
```

```python

import numpy as np
from scripts.DistanceHelper import get_distance, DistanceType
from numpy import dot
from numpy.linalg import norm

number_of_rows = frequency_term_document_matrix.shape[0]
number_of_columns = frequency_term_document_matrix.shape[1]
euclidean_distance_matrix = np.zeros((number_of_rows, number_of_rows))
cosine_similarity_matrix = np.zeros((number_of_rows, number_of_rows))

for i, element_0 in frequency_term_document_matrix.iterrows():
    for j, element_1 in frequency_term_document_matrix.iterrows():
        euclidean_distance_matrix[i, j] = get_distance(
            element_0, element_1, number_of_columns, DistanceType.EUCLIDEAN
        )
        cosine_similarity_matrix[i, j] = dot(element_0, element_1)/(norm(element_0)*norm(element_1))
        
print(euclidean_distance_matrix)
print()
print(cosine_similarity_matrix)
```
