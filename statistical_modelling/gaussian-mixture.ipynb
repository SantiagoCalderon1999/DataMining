```python
import pandas as pd

data = pd.read_csv('data/house-prices.csv')
data = data.loc[:, (data.columns == 'SqFt') | (data.columns == 'Price')]
data
```

```python
import matplotlib.pyplot as plt

x = data['Price']
y = data['SqFt']

plt.scatter(x, y)
plt.show()
```

```python
import sklearn.mixture as mixture

gmm = mixture.GaussianMixture(n_components=2, covariance_type='spherical')
gmm.fit(data)

gmm.means_
```

```python
labels = gmm.predict(data)
```

```python
plt.scatter(x, y, c = labels)
plt.plot()
```

```python
import sklearn.metrics as metrics

for i in range(2, 10):
    gmm = mixture.GaussianMixture(n_components=i, covariance_type='spherical')
    gmm.fit(data)
    ll_score = gmm.score(data)
    labels = gmm.predict(data)
    SC = metrics.silhouette_score(data, labels, metric = 'euclidean')
    CH = metrics.calinski_harabasz_score(data, labels)
    print(f'\n\n\nK: {i}')
    print(f'Log-likelihood score: {ll_score}')
    print(f'Silhouette score: {SC}')
    print(f'Calinski Harabasz score: {CH}')
```
